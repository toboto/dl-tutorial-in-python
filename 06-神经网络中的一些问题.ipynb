## SGD存在的一些问题

### 收敛的速度慢或路径不适合

- Momentum

	- 引入速度的概念
	- 每次迭代时更新量包含上次的速度
	- 迭代路径非常平滑

- AdaGrad

	- 随着学习深入降低学习率
	- 通过除以累计参数的平方和降低学习率

- Adam

	- 融合了Momentum和AdaGrad
	- 通过设置超参调整学习收敛速度

## 权重的初值

### 不能设置为0

### 隐藏层激活值分布

- 使用sigmoid函数容易出现梯度消失
- 分布过于集中网络的表现力不足

## Batch Norm层让激活值具有适当的广度

### 按照正态分布进行归一化

### 神经网络具有更快的收敛速度

### Batch Norm也可以以层的形式来表示

## 超参的选择

### 如何分配训练序列和测试序列

- 只能用训练序列来评价超参的选择，不要使用测试序列
- 训练序列应具有良好的随机性，保证网络的泛化能力

### 如何选择超参

- 实践中，随机选择具有良好的效果

	- 搜索范围可以大一些

- 理论上，可以参考贝叶斯最优化

## 过拟合

### 原因

- 模型参数多，表现力强
- 训练数据少

### 表现

- 对于训练序列拟合度非常好，甚至接近100%
- 对于测试数据则表现有明显差距

### 解决方法

- 权值衰减

	- 在损失函数上增加权值的L2范数乘以正则化强度
	- 对矩阵W的梯度函数要加上W乘以正则化强度
	- 增加了权值矩阵的L2范数后抑制了权值矩阵过大

- Dropout

	- 训练时，在神经网络中随机删除一些神经元
	- 测试时，乘上有效节点的比例
	- 随机删除节点抑制了正向传播
	- 也可以理解Dropout是达到了集成学习的效果
